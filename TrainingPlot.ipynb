```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('history')

# Set pandas display option to show full decimal values
pd.options.display.float_format = '{:.6f}'.format

# Extract the data for plotting
loss = data['loss']
val_loss = data['val_loss']
mean_absolute_error = data['mean_absolute_error']
val_mean_absolute_error = data['val_mean_absolute_error']
mean_squared_error = data['mean_squared_error']
val_mean_squared_error = data['val_mean_squared_error']

# Create the plot
plt.figure(figsize=(10, 6))
# plt.plot(loss, label='Loss')
# plt.plot(val_loss, label='Validation Loss')
# plt.plot(mean_absolute_error, label='Mean Absolute Error')
# plt.plot(val_mean_absolute_error, label='Validation Mean Absolute Error')
plt.plot(mean_squared_error, label='Mean Squared Error')
plt.plot(val_mean_squared_error, label='Validation Mean Squared Error')

# Add labels and legend
plt.xlabel('Epochs')
plt.ylabel('Values')
plt.title('Loss and Metrics Over Epochs')
plt.legend()

# Format y-axis tick labels in decimal form
plt.gca().yaxis.set_major_formatter(mticker.ScalarFormatter(useMathText=True))
plt.gca().ticklabel_format(style='plain', axis='y')

# Show the plot
plt.show()
```
